13. What is the difference between for and while loops?

Ans. Loops are used in programming language to execute repetitive tasks or execute same code block multiple times.
“For loop” iterates over a sequence (like a list, tuple, or string) and executes the block for each item.It is used when the total number of iterations is known and finite. “While loop”  repeats the block as long as a given condition is true. It is used when the number of iterations depends on dynamic conditions.

14. Describe a scenario where a while loop is more suitable than a for loop.

Ans. Consider an application that repeatedly asks a user to enter a command or a color until the user types "exit" or "quit." Since the number of inputs is unknown ahead of time, the program cannot rely on a for loop with a predetermined number of iterations. Instead, a while loop checks continuously whether the input meets the exit condition, and only then stops running. This approach fits perfectly with situations where the loop must continue "while" a condition is true, such as waiting for user input or sensor data until a stopping criterion is achieved.

### 6. Write a function to demonstrate the behavior of mutable and immutable arguments.

```python
def Mutable_immutable_Fn(im_arg, m_arg):
    print("Before modification:")
    print("Immutable arg:", im_arg)
    print("Mutable arg:", m_arg)

    # Modify immutable argument
    im_arg += 10

    # Modify tmutable argument
    m_arg.append(100)

    print("\nPost-modification using function:")
    print("Immutable arg:", im_arg)
    print("Mutable arg:", m_arg)


immutable_value = 7
mutable_value = [4, 7, 17]

Mutable_immutable_Fn(immutable_value, mutable_value)

print("\nOutside Function call:" )
print("Immutable arg:", immutable_value)
print("Mutable arg:", immutable_value)

```

### 7. What is the difference between " == " and "is" operators in Python?

#### "==" is a Comparison operator. They evaluate whether the content or data held by two variables is the same.

```python
# Example

a = "Suresh"
b = "Ramesh"

if a == b:
  print("a and b are same values")
else:
  print("a and b is not same")
```

#### "is" is an Identity operators. They evaluate whether two variables refer to the exact same object in memory.

```python
# Example
a = [1,2,3]
b = [1,2,3]
c = a

if a is b:
  print("a and b are the same object")
else:
  print("a and b are not the same object")

if a is c:
  print("a and c are the same object")
else:
  print("a and c are not the same object")
```

### 13. Write a Python program to find the sum of all even numbers between 1 and 50.

```python
start = int(input("Enater start value:"))
end = int(input("Enater end value:"))


sum = 0
for num in range(start, end+1):
  if num%2 == 0:
    sum += num
print(f"Sum of even numbers between {start} and {end} is {sum}")
```

### 14. Write a program to reverse a string using a while loop.

```python
s = 'was it a cat I saw'
e = len(s)-1
i = e
while i <= e:
  print(s[i], end="")
  i -= 1
  if i == -1:
    break
```

### 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
num = int(input("Enter a number:"))
i=1
f = 1
while i <= num:
  f *= i
  i +=1

print(f"Factorial of {num} = {f}")
```
