1. What is the difference between a function and a method in Python?

Ans. Key difference are:

1.	Functions are defined independently while methods are defined under a class.

10. What is the difference between `map()', 'reduce()', and `filter()` functions in Python?

Ans. Map(), resuce() and filter() all these 3 functions are applied on iterables and yield results of the function on elements of the iterables. Only difference is map() is used to apply the function all elements of the iterables and returns all transformed elements of the iterables. Reduce() as name suggested, aggregate the values of iterables. Whereas, filter() is used to select elements from the iterables based on condition provided in filter function.

Example:

from functools import reduce
nums = [1, 2, 3, 4]
print(list(map(lambda x: x**2, nums)))
print(reduce(lambda x, y: x + y, nums))
print(list(filter(lambda x: x % 2 == 0, nums)))

### 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.


```python
def fib(n):
    a, b = 0, 1
    for i in range(n):
        yield a
        a, b = b, a + b

print(list(fib(8)))
```

### 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.


```python
def powerof_2(x):
    for i in range(x+1):
        yield 2**i
    
print(list(powerof_2(3)))
    
```

### 7. Implement a generator function that reads a file line by line and yields each line as a string.


```python
def read_lines(path):
    with open(path, "r") as f:
        for line in f:
            yield line.strip("\n") 

            
for line in read_lines("temp.txt"):
    # process each line
    print(line)
```
