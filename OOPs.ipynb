2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.


```python
import abc

class Shape:
    
    @abc.abstractmethod
    def area(self):
        pass
    
class Circle(Shape):    
    def area(self):
        print("Area of Circle: pi*r**2")
        
class Rectangle(Shape):
    def area(self):
        print("Area of Rectagle: height*width")
        
r = Rectangle()
r.area()

c = Circle()
c.area()
        
```

5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance
        
    def deposit(self, amount):
        self.__balance = self.__balance + amount
        
    def withdraw(self, amount):
        if self.__balance >= amount:
            self.__balance = self.__balance - amount
            return True
        else:
            return False
        
    def check_balance(self):
        return self.__balance
    
acc1 = BankAccount(10000)
acc1.deposit(1000)
acc1.withdraw(750)
acc1.check_balance()
```

10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __add__(self,other):
        return Vector(self.x + other.x, self.y + other.y )
        
v1 = Vector(11,12)
v2 = Vector(14,15)
v3 = v1+v2
print(v3.x, v3.y)
```

```python
class Rectangle:
    def __init__(self):
        self.length = 0
        self.height = 0
        
    def set_dimensions(self,length,height):
        self.length = length
        self.height = height
        if self.length < 0 or self.height < 0:
            print("ValueError: Length and Height must be positive")
        
    def area(self):
        if self.length >= 0 or self.height >= 0:
            return self.length*self.height
    
Rec1 = Rectangle()
Rec1.set_dimensions(5,7)
print(Rec1.area())

Rec2 = Rectangle()
Rec2.set_dimensions(-5,5)
            
```

14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.

```python
class Employee:
    def __init__(self,hours,hourly_rate):
        self.hours = hours
        self.hourly_rate = hourly_rate
        
    def calculate_salary(self):
        return self.hourly_rate*self.hours
        
class Manager(Employee):
    def __init__(self,hours,hourly_rate, bonus):
        super().__init__(hours,hourly_rate)
        self.bonus = bonus
    
    def add_bonus(self):
        base_salary = super().calculate_salary()
        return base_salary+self.bonus
emp1 = Employee(40,30) 
print(emp1.calculate_salary())
manager1 = Manager(40,30,700)
print(manager1.add_bonus())
```

16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.

```python
import abc 

class Animal():
    @abc.abstractmethod
    def sound(self):
        pass


class Cow(Animal):
    def sound(self):
        return "Hamba"


class Sheep(Animal):
    def sound(self):
        return "Baaa"


c = Cow()
s = Sheep()
print(c.sound())
print(s.sound())
```
