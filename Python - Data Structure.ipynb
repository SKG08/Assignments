1. What are data structures, and why are they important?

2. Explain the difference between mutable and immutable data types with examples.
Ans. In case of mutable  data types, user can add, remove or modify the elements of the variables. Example – List and Set. Whereas, elements of immutable data types can be added, removed or altered after the data type is created. Example- Tuples and String.

7. How do tuples ensure data integrity in Python?
Ans. Tuples is immutable and can not be modified after creation. This characteristic of tuple ensures any data stored in can not be mistakenly or intentionally  modified and maintain the integrity of the data. 

8. What is a hash table, and how does it relate to dictionaries in Python?
Ans. Python dictionaries store key-value pairs where Python internally uses a hash function to process the keys. This hash function generates a unique hash value for each unique key. This unique hash value is used to determine where the key-value pair will be stored in the dictionary's underlying data structure. Thus,  Python, dictionaries are an implementation of hash tables.

12. Describe a scenario where using a tuple would be preferable over a list.
Ans. Tuples are immutable collection of data whereas, list is mutable. Hence, when the data integrity is important then tuple is obvious choice over list.

13. How do sets handle duplicate values in Python?
Ans. In Python, sets automatically handle duplicate values by not allowing any duplicates to exist within the set. If a new elements is  added to a set that is already present, the set ignores this addition, effectively keeping only one instance of each element. This behavior is due to sets being collections of unique elements based on their hash values.

14. How does the "in" keyword work differently for lists and dictionaries?
Ans. In case of list, the "in" keyword checks membership by  looking up elements sequentially to see if the specified value is present. For dictionaries, the "in" keyword checks for the presence of a key, not a value. 
, "in" on a list searches for the existence of a value , while "in" on a dictionary checks for the existence of a key using hashing. 

15. Can you modify the elements of a tuple? Explain why or why not.
Ans. Elements of a tuple can to be changed in Python because tuples are immutable. This means that once a tuple is created, its contents  and their order cannot be changed, added to, or removed. 
However, if a tuple contains mutable objects (such as a list), those internal mutable objects can be changed, but the tuple itself still references the same objects in the same order. 
Example:
t = (1,2, [3,4])
print(t)
t[2][1] = 5
print(t)

Output:
(1, 2, [3, 4])
(1, 2, [3, 5])

17. Describe the time complexity of accessing elements in a dictionary.
Ans. The time complexity of accessing elements in a dictionary is generally O(1), meaning it allows constant-time access on average. This efficiency is because dictionaries use a hash table data structure internally, which calculates a hash value from the key to directly locate the memory address of the value. 

18. In what situations are lists preferred over dictionaries?
Ans. List are simple form of data structure and use less memory. List is preferred over dictionaries when data is stored and retrieved in a sequential manner using position. 

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
Ans. Dictionaries are considered unordered because the way they store data does not keep items in a specific sequence like a list does. Instead of remembering the order in which items were added, dictionaries focus on organizing data for very fast access using keys. Dictionaries u can't retrieve items by position or index; the order can seem random or unpredictable.

20. Explain the difference between a list and a dictionary in terms of data retrieval.
Ans. In a list, you retrieve items by their position or order. Each item has a numeric index starting from zero, and you use this index to get the item. Whereas, in dictionary, data is retrieved by a unique key rather than position. Instead of numeric indexes, each item is stored with a key that is used to get the corresponding value quickly. Data retrieving is faster in case of dictionary compared to list as it does not need to follow any indexing or sequence.
