2. How does broadcasting work in NumPy?

Ans. NumPy broadcasting enables arithmetic operations between arrays of different shapes by virtually expanding the smaller array to match the larger one, without creating a separate array.
Numpy Broadcasting works on following rules.
1.	For each dimension, the sizes must either be same or one of them is 1
2.	Size in each dimension of the output shape is maximum of the input sizes in that dimension
3.	if an input has a dimension size of 1, the first data entry in that dimension is used for all calculations along that dimension.
4.	If in any dimension the sizes disagree and neither is equal to 1, a ValueError: operands could not be broadcast together exception is raised, and the operation cannot be performed.

15. How does NumPy handle multidimensional arrays?

Ans. The actual data of a multidimensional array is stored in a flat, one-dimensional contiguous memory block. NumPy doesn't physically rearrange the data to create dimensions. Instead, the ndarray object uses metadata, primarily its shape (a tuple of integers defining the size of each dimension) and strides (the number of bytes to skip to move to the next element along a specific dimension), to provide a multidimensional view of the data.

17. Explain the difference between apply() and map() in Pandas.

Ans. Pandas apply() and map() serve distinct purposes for function application on data structures, with map() operating element-wise exclusively on Series objects and apply() providing flexible row- or column-wise application on both Series and DataFrames.

18. What are some advanced features of NumPy?

Ans Key Advanced Features of Numpy:
·	Broadcasting: Automatically expands arrays of different shapes for element-wise operations without explicit looping, optimizing memory and speed.
·	Universal Functions (ufuncs): Vectorized, optimized functions like np.add or np.sin that apply element-wise across arrays with automatic type casting and broadcasting support.
·	Advanced Indexing: Supports fancy indexing (arrays of indices), boolean masking, and slicing for precise, non-contiguous data selection without copies.
·	Array Views and Manipulation: Zero-copy views via slicing, reshaping with np.reshape, stacking/splitting (hstack, vstack), and transpose for flexible data restructuring.
·	Linear Algebra (linalg): Built-in solvers for eigenvalues, SVD, determinants, matrix inverses, and linear systems via np.linalg.

21. Why is NumPy's array slicing faster than Python's list slicing?

Ans. NumPy array slicing is faster than Python list slicing primarily because it returns a view (zero-copy reference to the original contiguous memory block) rather than creating a full data copy, avoiding expensive memory allocation and data duplication.

### 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
data = pd.DataFrame({"x1": [22,45,67,85],
                     "x2": [34,30,25,17],
                     'x3': [11,15,8,3]
})
mat1 = data.corr()

plt.figure(figsize=(12, 3))
sns.heatmap(mat1, annot=True, cmap='coolwarm', center=0, square=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
